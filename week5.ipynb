> Set index of data to hn numbers 

```python
df = pd.read_csv('https://raw.githubusercontent.com/plenoi/Clinic/master/ultima_all_clean.csv', sep=',', header = 0)
df = df.set_index('hn')
df
```

> Remove missing values

```python
df_clean = df
df_clean
```

```python
df_clean = df_clean.dropna(0,'any')
df_clean
```

> Remove these unnecessary columns hiv, size, utmet, vgmet, surgery, pchemo 

```python
df_clean  = df_clean .drop('hiv',1)
df_clean  = df_clean .drop('size',1)
df_clean  = df_clean .drop('utmet',1)
df_clean  = df_clean .drop('vgmet',1)
df_clean  = df_clean .drop('surgery',1)
df_clean  = df_clean .drop('pchemo',1)
df_clean
```

> Change age into 2 values 'Young' (age <= 20) and 'Old' (age > 20)


```python
old = df_clean['age'] > 20
df_clean['age'] = old
```

```python
df_clean['age'] = df_clean['age'].replace(False,'Young') 
df_clean['age'] = df_clean['age'].replace(True,'Old')
df_clean
```

> Separate data into groups based on pmmet values 


> Separate data into groups based on pmmet values 
for each group, randomly choose 10 data

```python
df_clean2= df_clean.groupby('pmmet').sample(n=10, random_state=1)
df_clean2.sort_values(by=['pmmet'])
```
